# Spark Quickstart\n
Notebook di esempio collegato al cluster Spark Docker Compose.

```python
from pyspark.sql import SparkSession\n
\n
spark = (\n
    SparkSession.builder\n
    .appName('quickstart-notebook')\n
    .master('spark://spark-master:7077')\n
    .getOrCreate()\n
)\n
\n
spark
```

```python
df = spark.read.text('/home/jovyan/work/data/input.txt')\n
df.show(truncate=False)
```

```python
from pyspark.sql.functions import explode, split, lower, col\n
\n
word_counts = (\n
    df.select(explode(split(lower(col('value')), r'\\s+')).alias('word'))\n
      .where(col('word') != '')\n
      .groupBy('word')\n
      .count()\n
      .orderBy(col('count').desc(), col('word').asc())\n
)\n
\n
word_counts.show(truncate=False)
```
